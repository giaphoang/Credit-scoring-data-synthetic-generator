```python
from sdv.single_table import CTGANSynthesizer
from sdv.metadata import SingleTableMetadata

metadata = SingleTableMetadata()
metadata.detect_from_dataframe(data=df)

# Drop columns with high cardinality that are causing memory issues
columns_to_drop = [
    'Name',
    'SSN',
    'Annual_Income',
    'Num_of_Loan',
    'Type_of_Loan',
    'Changed_Credit_Limit',
    'Outstanding_Debt',
    'Credit_History_Age',
    'Amount_invested_monthly',
    'Monthly_Balance',
    'ID',
    'Customer_ID' # Also drop ID columns as they are not useful for synthesis
]

df_reduced = df.drop(columns=columns_to_drop)

metadata_reduced = SingleTableMetadata()
metadata_reduced.detect_from_dataframe(data=df_reduced)


synthesizer = CTGANSynthesizer(
    metadata_reduced,
    enforce_rounding=False,
    epochs=500,
    verbose=True
)

synthesizer.fit(df_reduced)

synthesizer.save(filepath='credit_scoring_model.pkl')
```
